# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 3 + 1 * x  + np.random.normal(0, 1, size=100)

plt.figure(figsize=(10, 8))
plt.scatter(x, y, c='b', marker='x')
plt.show()
```
