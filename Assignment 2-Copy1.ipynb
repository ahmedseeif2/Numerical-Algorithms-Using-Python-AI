Deff methods 



```python
def bisection(a,b,epsilon):
    count = 0
    
    while abs(a-b)>epsilon:
        
        g = (a+b)/2
        if f(a)*f(g)<0:
            b = g
        else:
            a = g
        count +=1
    
    return g , count

def Fixed_point(p,epsilon):
    count = 0
    p1 = p
    p2 = g(p1)
    
    while abs(p1-p2)>epsilon:
        p1 = p2
        p2 = g(p1)
        count +=1
    
    return p2 , count

def Newton(p,epsilon):
    count = 0
    x1 = p
    x2 = x1-(f(x1)/df(x1))
    
    while abs(x1-x2)>=epsilon:
        x1 = x2
        x2 = x1-(f(x1)/df(x1))
        count +=1
    
    return x2,count
```
