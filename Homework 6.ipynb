```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def Q(x):
    x1 = x[0]
    x2 = x[1]
    
#The Quadratic function will be as follow


    return 0.5 * x1**2 + 0.5*x2**2 - x1*x2 
```

```python
LB = -4
UB = 4
x1 = np.linspace(LB, UB, 50)
x2 = np.linspace(LB, UB, 50)
X, Y = np.meshgrid(x1, x2)
Z = Q([X, Y])

#Plot 2D Contour


fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(121)
ax1.contour(X, Y, Z, levels=30)
#Plot the Surface
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z)
plt.show()
```

```python
from scipy.optimize import minimize

def F(x):
    return (x[1]**2 + x[0]- 11)**2 + (x[0]**2 + x[1]- 7)**2

x0 = np.array([4, -4])
res = minimize(F, x0, method='Powell', options={'disp':True, 'return_all':True})
xs = res.allvecs


LB = -5
UB = 5
x = np.linspace(LB, UB, 50)
y = np.linspace(LB, UB, 50)
X, Y = np.meshgrid(x, y)

Z = F([X, Y])

ax1.contour(X, Y, Z, levels=20)

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)
ax1.contour(X, Y, Z, levels=20)
ax1.scatter([x[0] for x in xs], [x[1] for x in xs], c=list(range(len(res.allvecs))), marker='x')
all_points= np.array(res.allvecs)
plt.plot(all_points[:,0], all_points[:,1])

plt.show()

print(res.x)
```

```python
from scipy.optimize import minimize

def F(x):
    return (x[0]**2 + x[1]- 11)**2 + (x[1]**2 + x[0]- 7)**2

def jacobian(x):
    return np.array([ 4*x[0]*(x[0]**2 + x[1]- 11) + 2*(x[1]**2 + x[0]- 7), 2*(x[0]**2 + x[1]- 11) + 4*x[1]*(x[1]**2 + x[0]- 7) ])


x0 = np.array([4, -4])
res = minimize(F, x0, method='CG', jac=jacobian , options={'disp':True, 'return_all':True})
xs = res.allvecs


LB = -5
UB = 5
x = np.linspace(LB, UB, 50)
y = np.linspace(LB, UB, 50)
X, Y = np.meshgrid(x, y)

Z = F([X, Y])

ax1.contour(X, Y, Z, levels=20) 

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)
ax1.contour(X, Y, Z, levels=20)
ax1.scatter([x[0] for x in xs], [x[1] for x in xs], c=list(range(len(res.allvecs))), marker='x')
all_points= np.array(res.allvecs)
plt.plot(all_points[:,0], all_points[:,1])
plt.show()


print("Number of Itrations =  ",res.nit)
print("Found minimum = " ,res.x)
print("Number of function evaluations = " ,res.nfev)
```
